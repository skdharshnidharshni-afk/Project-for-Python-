```python
import matplotlib.pyplot as plt
expenses=[500,900,800]
categories=["food","travel","shopping"]
plt.pie(expenses,labels=categories)
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may"]
sales=[10000,12000,15000,13000,18000]
plt.plot(months,sales,
         color="green",
         marker="o",
         linestyle="--")
plt.title("sales treand")
plt.xlabel("month")
plt.ylabel("sales")
plt.show
```

```python
import matplotlib.pyplot as plt
products=["laptop","moblie","tablet"]
sales=[50,80,90]
plt.bar(products,
        sales,
        color=["red","blue","green"])
plt.title("products sales")
plt.show()
```
