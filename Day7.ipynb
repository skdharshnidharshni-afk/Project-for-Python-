```python
import pandas as pd

employees={
"Emp_ID":[101,301,501,601,701],
"Name":["priya","manoj","gaja","suba","madhu"],
"Department":["IT","HR","FINANCE","MANAGER","IT"],
"Salary":[45000,89000,45600,12300,54300],
"Experience":[2,3,6,8,9]
}

df=pd.DataFrame(employees)

print("Employee Dataset")
print(df)

print("\n Average Salary")
print(df["Salary"].mean())

print("\n Highest Salary")
print(df["Salary"].max())

print("\n Employees in HR")
print(df[df["Department"]=="HR"])

print("\n Empolyees with Experience>3")
print(df[df["Experience"]>3])

print("\n Empolyees count by department")
print(df["Department"].value_counts())
```
